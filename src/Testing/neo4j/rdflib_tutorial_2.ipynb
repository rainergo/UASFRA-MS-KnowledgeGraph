```python
import rdflib
from rdflib.graph import Graph
# import neosemantics

# Load the ontology from the Turtle file
graph = Graph()
graph.parse(path_to_ontology, format='ttl')

# Convert the ontology to Cypher queries
cypher_queries = []
for triple in graph.triples((None, None, None)):
    subject, predicate, object_ = triple

    # Handle annotations
    if isinstance(predicate, rdflib.URIRef) and predicate.startswith('http://www.w3.org/2000/01/rdf-schema#'):
        print('isinstance(predicate, rdflib.URIRef):', isinstance(predicate, rdflib.URIRef))
        if predicate == rdflib.RDFS['subClassOf']:
            cypher_queries.append(f'MATCH (c:{subject}) MATCH (p:{object_}) CREATE (c)-[:SUBCLASS_OF]->(p)')
        else:
            cypher_queries.append(f'MATCH (n:{subject}) SET n.{predicate} = {object_}')
    else:
        cypher_queries.append(f'MATCH (s:{subject}) MATCH (o:{object_}) CREATE (s)-[:{predicate}]->(o)')

# Connect to Neo4j and execute the Cypher queries
# db = neosemantics.Connection('bolt://localhost:7687')
# for cypher_query in cypher_queries:
#     db.run_cypher(cypher_query)

for q in cypher_queries:
    print('QUERY:\n', q)
    print('-----------')
```
